```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

# Merge text from two columns
questions = list(ques_df['question1']) + list(ques_df['question2'])

# Handle missing values by replacing NaN with empty strings
questions = [str(q) if not pd.isna(q) else '' for q in questions]

# Create CountVectorizer with max_features
cv = CountVectorizer(max_features=3000)

# Transform the merged text into numerical features
X = cv.fit_transform(questions).toarray()

# Split the resulting array into two parts for q1_arr and q2_arr
split_index = len(ques_df)  # Assuming ques_df is your DataFrame
q1_arr, q2_arr = X[:split_index], X[split_index:]
```

```python
temp_df1 = pd.DataFrame(q1_arr, index= ques_df.index)
temp_df2 = pd.DataFrame(q2_arr, index= ques_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df.shape
```
